### 83. Remove Duplicates from Sorted List

Given a sorted linked list, delete all duplicates such that each element appear only once.

```python
    def deleteDuplicates(head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        p = ListNode(None)
        p.next = head
        values = set()
        def deldup(head, values):
            if head == None or head.next == None:
                return 
            if head.next.val not in values:
                values.add(head.next.val)
                head = head.next
            else:
                head.next = head.next.next
            deldup(head, values)
            return
        deldup(p, values)
        return p.next
```

```python
class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        if head == None:
            return None
        index = head
        point = head.next
        while point is not None:
            if point.val != index.val:
                index.next = point
                index = point
            else:
                index.next = None
            point = point.next
        return head
```

### 160. Intersection of Two Linked Lists (the two pointer method is so cool!)

Write a program to find the node at which the intersection of two singly linked lists begins.

```python
class Solution(object):   
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        # headA go faster
        if headA == None or headB == None:
            return None
        pa = headA
        pb = headB
        while pa or pb:
            if pa == None:
                pa = headB
            if pb == None:
                pb = headA
            if pa == pb:
                return pa
            pa = pa.next
            pb = pb.next
        return None
```

```python
class Solution(object):
    def getLen(self,head):
        counter = 0
        while head:
            counter += 1
            head = head.next
            
        return counter
    
    def getNewHead(self,head,idx):
        counter = 0
        
        while counter < idx:
            head = head.next
            counter += 1
        
        return head
    
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        
        lenA = self.getLen(headA)
        lenB = self.getLen(headB)
        
        if lenA > lenB:
            idxA = lenA - lenB
            nodeA = self.getNewHead(headA,idxA)
            nodeB = headB
        else:
            idxB = lenB - lenA
            nodeB = self.getNewHead(headB,idxB)
            nodeA = headA
        
        
        while nodeA and nodeB:
            if nodeA == nodeB:
                return nodeA
            nodeA = nodeA.next
            nodeB = nodeB.next
        else:
            return 
```

### 445. Add Two Numbers II

You are given two non-empty linked lists representing two non-negative integers. The most significant digit comes first and each of their nodes contain a single digit. Add the two numbers and return it as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

```python
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        s1, s2 = [], []
        p1, p2 = l1, l2
        while p1:
            s1.append(p1.val)
            p1 = p1.next
        
        while p2:
            s2.append(p2.val)
            p2 = p2.next
        prev = None
        carry = 0
        while s1 or s2 or carry!=0:
            if not s1 and not s2:
                tmp = carry
            elif not s1:
                tmp = s2.pop() + carry
            elif not s2:
                tmp = s1.pop() + carry
            else:
                tmp = s2.pop() + s1.pop() + carry
            cur = tmp%10
            carry = tmp/10
            newnode = ListNode(cur)
            newnode.next = prev
            prev = newnode
        return prev
```

### 24. Swap Nodes in Pairs (well done!)

Given a linked list, swap every two adjacent nodes and return its head.

```python
class Solution(object):
    def swapPairs(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        p = ListNode(0)
        p0 = p
        p.next = head
        while p.next and p.next.next:
            mid = p.next
            end = p.next.next
            p.next, mid.next, end.next = mid.next, end.next, mid
            p = mid            
        return p0.next
```

### 148. Sort List

Sort a linked list in O(n log n) time using constant space complexity.

```python
class Solution(object):    # quick sort
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        # try quick sort
        def get_mid(head):
            s = f = head
            while f and f.next:
                s, f = s.next, f.next.next
            return s.val
        
        def partition(head, val):
            # partition around val
            fake_l = l = ListNode(0)
            fake_m = m = ListNode(0)
            fake_r = r = ListNode(0)
            cur = head
            while cur:
                if cur.val < val:
                    l.next = cur
                    l = l.next
                elif cur.val == val:
                    m.next = cur
                    m = m.next
                else:
                    r.next = cur
                    r = r.next
                cur = cur.next
            l.next, m.next, r.next = None, None, None
            return fake_l.next, fake_m.next, fake_r.next
        
        def get_tail(head):
            cur = head
            while cur.next:
                cur = cur.next
            return cur
        
        def combine(l, m, r):
            # l, m, r could be None
            fake = ListNode(0)
            fake.next = l
            t = get_tail(fake)
            t.next = m
            t = get_tail(t)
            t.next = r
            return fake.next
        
        if not head or not head.next:
            return head
        l, m, r = partition(head, get_mid(head))
        l = self.sortList(l)
        r = self.sortList(r)
        return combine(l, m, r) 
```

### 725. Split Linked List in Parts

Given a (singly) linked list with head node root, write a function to split the linked list into k consecutive linked list "parts".

The length of each part should be as equal as possible: no two parts should have a size differing by more than 1. This may lead to some parts being null.

The parts should be in order of occurrence in the input list, and parts occurring earlier should always have a size greater than or equal parts occurring later.

Return a List of ListNode's representing the linked list parts that are formed.

Examples 1->2->3->4, k = 5 // 5 equal parts [ [1], [2], [3], [4], null ]

```python
class Solution(object):
    def splitListToParts(self, root, k):
        """
        :type root: ListNode
        :type k: int
        :rtype: List[ListNode]
        """
        nodelist = [root]
        lenlist = []
        total_len = 0
        cal = root
        while cal:
            total_len += 1
            cal = cal.next
        avglen = int(total_len/k)
        res = total_len - k*avglen
        for i in range(k):
            res -= 1
            if res >= 0:
                lenlist.append(avglen+1)
            else:
                lenlist.append(avglen)
            
        n = k
        while root:
            lenlist[n-k] -= 1            
            if lenlist[n-k] == 0:
                k -= 1
                temp = root.next
                root.next = None
                root = temp
                if root != None:
                    nodelist.append(root)
            else:
                root = root.next
        if len(nodelist) < n:
            nodelist.extend([None for i in range(n - len(nodelist))])
        return nodelist
```

```python
class Solution(object):
    def splitListToParts(self, root, k):
        """
        :type root: ListNode
        :type k: int
        :rtype: List[ListNode]
        """
        node = root
        n = 0
        ans = []
        while node:
            n += 1
            node = node.next
        num, rest = divmod(n, k)
        cur = root
        for i in range(k):
            head = cur
            for j in range(num + (i < rest) - 1):
                if cur:
                    cur = cur.next
            if cur:
                cur.next, cur = None, cur.next
            ans.append(head)
        return ans
```

You may not modify the values in the list's nodes, only nodes itself may be changed.

```python
    def reorderList(head):  # run time exceeds the limit
        """
        :type head: ListNode
        :rtype: void Do not return anything, modify head in-place instead.
        """
        nodelist = {}
        temp = head
        i = 0
        while temp:
            nodelist[i] = temp
        k = 0
        n = len(nodelist)
        while k < int((n+1)/2) -2:
            head.next = nodelist[n-k]
            head.next.next = nodelist[k+1]
            head = head.next.next
            k += 1
        # approach the middle
        if n % 2 == 1:
            head.next = None
            return
        else:
            head.next = nodelist[n-k]
            head.next.next = None
            return 
```

```python
class Solution:
    def reorderList(self, head):
        """
        分成两个链表，前steps_to_next个元素为L1，头尾为h1,t1
        后面的所有元素，头尾为h2,t2
        如： 链表1~7, h1 = 1 t1 = 4 h2 = 5 t2 = 7
        """
        
        if not head or not head.next:
            return
        
        # get N, the length of the List, which costs N and store it in 'count'
        l = head
        r = head
        count = 1 
        while r.next:
            r = r.next
            count += 1

        # get the first half of the List,and store its head & tail in h1 and t1
        h1 = head
        t1 = head
        steps_to_next_half = count // 2 - 1 if count % 2 == 0 else count // 2
        for i in range(steps_to_next_half):
            t1 = t1.next
        h2 = t1.next
        # we have got the next half's tail
        t2 = r
        

        # 掐断前半部分，并开始翻转后面链表
        # cut down connection between two Lists, then reverse the second-half List
        t1.next = None
        cur = h2
        # use dummy node
        dummy = ListNode(-1)
        pre = dummy
        while cur:
            next = cur.next
            cur.next = pre
            pre = cur
            cur = next
        h2.next = None
        
        # 划分完毕
        # Since the number of the first-half List is not less than the second-half List,
        # we reorder the final List depends on the second-half List.
        # (Notice: Now t2 is the head of the second-half List and h2 is the tail.)
        while t2:
            h1_n,t2_n = h1.next, t2.next
            t2.next = h1_n
            h1.next = t2
            h1, t2 = h1_n,t2_n
						
        return 
```

### 92. Reverse Linked List II

Reverse a linked list from position m to n. Do it in one-pass.

Note: 1 $\le$ m $\le$ n $\le$ length of list.

```python
class Solution(object):
    def reverseBetween(self, head, m, n):
        """
        :type head: ListNode
        :type m: int
        :type n: int
        :rtype: ListNode
        """
        p = ListNode(0)
        p.next = head
        current = p
        while n>= 0:
            if m == 1:
                temp0 = current
                current = current.next
            elif m == 0:
                temp1 = current
                pre = current
                current = current.next           
            elif m < 0:
                tempt = current.next
                current.next = pre
                pre = current
                current = tempt
            else:
                current = current.next
            m -= 1
            n -= 1
        temp1.next = current
        temp0.next = pre
        return p.next
```

```python
class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        p = ListNode(0)
        p.next = l1
        flag = 0
        while l1.next and l2.next:
            s = l1.val + l2.val + flag
            flag = int(s/10)
            l1.val = s % 10
            l1 = l1.next
            l2 = l2.next
        s = l1.val + l2.val + flag
        flag = int(s/10)
        l1.val = s % 10
        if l2.next == None and l1.next!= None:
            while l1.next:
                l1 = l1.next
                s = l1.val + flag 
                flag = int(s/10)
                l1.val = s % 10
        elif l1.next == None and l2.next != None:
            l2 = l2.next
            while l2:
                s = l2.val + flag
                flag = int(s/10)
                l1.next = ListNode(s % 10)
                l1 = l1.next
                l2 = l2.next
        if flag > 0:
            l1.next = ListNode(flag)
        return p.next
```

### 86. Partition List

Given a linked list and a value x, partition it such that all nodes less than x come before nodes greater than or equal to x.

You should preserve the original relative order of the nodes in each of the two partitions.

```python
class Solution(object):
    def partition(self, head, x):
        """
        :type head: ListNode
        :type x: int
        :rtype: ListNode
        """
        h1 = l1 = ListNode(0)
        h2 = l2 = ListNode(0)
        while head:
            if head.val < x:
                l1.next = head
                l1 = l1.next
            else:
                l2.next = head
                l2 = l2.next
            head = head.next
        l2.next = None
        l1.next = h2.next
        return h1.next
```
